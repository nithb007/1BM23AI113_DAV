```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("lab4.csv")
display(df)
```

```python
print(f"Mean population : {df['Population'].mean()}")
```

```python
print(f"Total GDP:{df['GDP'].sum()}")
```

```python
max1 = df['Population'].max()
min1 = df['Population'].min()
display(max1)
display(min1)
```

```python
mean1 = df['GDP'].mean()
summ1 = df['GDP'].sum()
max2 = df['GDP'].max()
min2 = df['GDP'].min()
display(mean1)
display(max2)
display(summ1)
display(min2)
```
